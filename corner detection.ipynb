```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load image and convert it to gray
img = cv2.imread("images/chess board.jpg")
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#plot Image 

plt.imshow(img_gray,cmap='gray')
```

```python
# corners detection 
img_gray = np.float32(img_gray)
corners = cv2.cornerHarris(img_gray,2,3,0.04)
corners = cv2.dilate(corners,None)

plt.imshow(corners,cmap='gray')
```

```python
#draw circle around detected corners 
threshold = 0.01 * corners.max() # setting threshold for corners 
for i in range(0,corners.shape[0]):
    for j in range(0,corners.shape[0]):
        if(corners[i,j]>threshold):
            cv2.circle(img,(i,j),1,(0,255,0),2)

plt.imshow(img)
```

```python
#try one more image 

img2_c = cv2.imread('images/waffle.jpg')
img2 = cv2.cvtColor(img2_c,cv2.COLOR_BGR2GRAY)
plt.imshow(img2,cmap='gray')
```
