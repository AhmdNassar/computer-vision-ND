In this Notebook I practice to remove green screen from image and replace it with any other background

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read images
img = cv2.imread('images/car_green_screen.jpg')
backGround = cv2.imread('images/sky.jpg')
#img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#backGround = cv2.cvtColor(backGround,cv2.COLOR_BGR2RGB)

ret , (ax1 , ax2 ) = plt.subplots(ncols=2 , nrows= 1 , figsize=[20,10])
ax1.set_title('Original Image')
ax1.imshow(img)
ax2.set_title('background')
ax2.imshow(backGround)
```

## create mask to remove green background


```python
# first we convert color to rgb 

#then we create mask and plot it
mask = cv2.inRange(img,np.array([0,230,0]),np.array([200,255,200]))
plt.imshow(mask , cmap='gray')
```

```python
#make copy from original image
masked_img = np.copy(img)
masked_img[mask!=0] = [0,0,0]
plt.imshow(masked_img)
```

```python
# then we resize background to equal front image 
backGround_croped  = backGround[0:img.shape[0],0:img.shape[1],:]
print(backGround_croped.shape,img.shape)
backGround_croped[mask == 0] = [0,0,0]
plt.imshow[backGround_croped]
```
