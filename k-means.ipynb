##### In this notebook I will implement k-means algorithm with opencv

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load image and convert it from GBR to RGB
img = cv2.imread('images/oranges.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

K means in cv2 expect image with shape [?,3] so we first reshape img

```python
reImg = img.reshape((-1,3))
```

```python
reImg = np.float32(reImg)
critiria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,100,.2)
ret , labels , center = cv2.kmeans(reImg,2,None ,critiria,1,cv2.KMEANS_RANDOM_CENTERS)
```

```python
center = np.uint8(center)
res = center[labels.flatten()]
res2 = res.reshape((img.shape))
```

```python
plt.imshow(res2)
```
